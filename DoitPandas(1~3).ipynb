### 데이터 프레임 다루기

```python
#데이터프레임 불린 추출
print(sci[sci['Age']>sci['Age'].mean()])
```

```python
#데이터프레임의 'Born', 'Died'열의 자료형을 Object에서 Datetime으로 변환
born_datetime = pd.to_datetime(sci['Born'], format = '%Y-%m-%d')
```

```python
died_datetime = pd.to_datetime(sci['Died'], format = '%Y-%m-%d')
```

```python
sci['born_dt'], sci['died_dt'] = born_datetime, died_datetime
sci
```

```python
sci['living_period'] = sci['died_dt'] - sci['born_dt']
sci
```

```python
#데이터 섞어주기 (random 매서드)
import random
print(sci['Age'])
random.seed(42) #없어도 섞어지긴 하는데 SettingWithCopyWarning에러 뜸
random.shuffle(sci['Age'])
print(sci['Age'])
```
